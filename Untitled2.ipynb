```python
def convert_years_cs(x):
    if x=='1-3 years':
        return 2 
    if x=='Less than 1 year':
        return 0.5
    if x=='More than 5 years':
        return 7
    if x=='3-5 years':
        return 4
    else:
        return 0
```

```python
def construct_year_index(x):
    if x['Current_Role'] in ['Cybersecurity Analyst', 'Database Administrator', 'Other (please specify)', 'Cloud Engineer','IT Support Specialist','AI Researcher','Machine Learning Engineer']:
        return 1
    elif x['Current_Role'] in ['Software Engineer', 'Game Developer', 'Front-End Developer', 'DevOps Engineer', 'Back-End Developer']:
        return 2
    elif x['Current_Role'] in ['Data Scientist', 'Product Manager', 'Project Manager', 'Mobile Developer', 'Systems Analyst', 'Robotics Engineer', 'Data Engineer', 'Network Engineer']:
        return 3
    elif x['Current_Role'] == 'Full-Stack Developer':
        return 4
```

```python
df['year_index'] = df.apply(lambda x: construct_year_index(x), axis=1)
```

```python
def construct_language_index(x):
    # Check if 'Programming_Languages' is a list and contains the specified language
    if (x['Current_Role'] == 'Software Engineer') & (x['Programming_Languages'] == 'C++'):
        return 1
    if (x['Current_Role'] == 'Network Engineer') & (x['Programming_Languages'] == 'C#' or x['Programming_Languages'] == 'C'):
        return 2
    if (x['Current_Role'] == 'Game Developer') & (x['Programming_Languages'] == 'Kotlin'):
        return 3
    if (x['Current_Role'] == 'Data Scientist') & (x['Programming_Languages'] == 'Swift'):
        return 4
    else:
        return 5

# Assuming df has a column 'Programming_Languages' with single language values
df['language_index'] = df.apply(lambda x: construct_language_index(x), axis=1)

# Show the updated dataframe with the new 'language_index' column
print(df)
```

```python
def construct_framework_index(x):
   if (x['Current_Role'] == 'AI Researcher') & (x['Frameworks'] == 'Flutter' or x['Frameworks'] == 'React' ):
        return 1
   if (x['Current_Role'] == 'Cloud Engineer') & (x['Frameworks'] == 'Nestjs'):
        return 2
   
   if (x['Current_Role'] == 'Data Scientist') & (x['Frameworks'] == 'Kali Linux'  or x['Frameworks'] == 'React Native'  or x['Frameworks'] == 'SwiftUI' ):
        return 4
   
   if (x['Current_Role'] == 'DevOps Engineer') & (x['Frameworks'] == 'Pandas'  or x['Frameworks'] == 'Nestjs'  or x['Frameworks'] == 'React' ):
        return 5
   if (x['Current_Role'] == 'Game Developer') & (x['Frameworks'] == 'Tensorflow'  ):
        return 6 
   if (x['Current_Role'] == 'Machine Learning Engineer') & (x['Frameworks'] == 'Angular' or x['Frameworks'] == 'Unity' ):
        return 7
   if (x['Current_Role'] == 'Mobile Developer') & (x['Frameworks'] == 'Burp Suite'  or x['Frameworks'] == 'PyTorch'  or x['Frameworks'] == 'Ruby on Rails' ):
        return 8
   if (x['Current_Role'] == 'Network Engineer') & (x['Frameworks'] == 'Vue.js'  or x['Frameworks'] == 'Apache Spark'  or x['Frameworks'] == 'Express.js' ):
        return 9
   if (x['Current_Role'] == 'Other (please specify)') & (x['Frameworks'] == 'React'):
        return 10
   if (x['Current_Role'] == 'Product Manager') & (x['Frameworks'] == 'Scikit-Learn'):
        return 11
   if (x['Current_Role'] == 'Project Manager') & (x['Frameworks'] == 'Next.js'):
        return 12
   if (x['Current_Role'] == 'Robotics Engineer') & (x['Frameworks'] == 'Flask' or x['Frameworks'] == 'Flask' ):
        return 13
   if (x['Current_Role'] == 'Software Engineer') & (x['Frameworks'] == 'Express.js' or x['Frameworks'] == 'Flutter' ):
        return 14
   if (x['Current_Role'] == 'Systems Analyst') & (x['Frameworks'] == 'Dask'):
        return 15
   else :
       return 16
df['framework_index'] = df.apply(lambda x: construct_framework_index(x), axis=1)
```

```python
def construct_devops_index(x):
   if (x['Current_Role'] == 'AI Researcher') and (x['DevOps_Tools'] == 'Git'):
        return 1
   if (x['Current_Role'] == 'Cloud Engineer') and (x['DevOps_Tools'] == 'GitLab CI/CD'):
        return 2
    
   if (x['Current_Role'] == 'Back-End Developer') and (x['DevOps_Tools'] == 'DigitalOcean' or x['DevOps_Tools'] == 'AWS'):
        return 3
   if (x['Current_Role'] == 'Cybersecurity Analyst') and (x['DevOps_Tools'] in ['DigitalOcean']):
        return 4
   
   if (x['Current_Role'] == 'Data Scientist') and (x['DevOps_Tools'] in ['Apache']):
        return 5
   
   if (x['Current_Role'] == 'Data Engineer') and (x['DevOps_Tools'] in ['Docker']):
        return 6
   if (x['Current_Role'] == 'DevOps Engineer') and (x['DevOps_Tools'] in ['CircleCI', 'GitLab CI/CD']):
        return 7
   
   if (x['Current_Role'] == 'Front-End Developer') and (x['DevOps_Tools'] in ['Ansible', 'Azure']):
        return 8
   
   if (x['Current_Role'] == 'Full-Stack Develope') and (x['DevOps_Tools'] in ['Other (please specify)', 'CircleCI']):
        return 9
       
   if (x['Current_Role'] == 'Game Developer') and (x['DevOps_Tools'] in ['Apache','AWS']):
        return 10
   if (x['Current_Role'] == 'Machine Learning Engineer') and (x['DevOps_Tools'] in ['Apache']):
        return 11
   if (x['Current_Role'] == 'Mobile Developer') and (x['DevOps_Tools'] in ['Other (please specify)', 'Azure']):
        return 12
   if (x['Current_Role'] == 'Network Engineer') and (x['DevOps_Tools'] in ['Google Cloud Platform', 'Kubernetes','Ansible']):
        return 13
   if (x['Current_Role'] == 'Product Manager') and (x['DevOps_Tools'] == 'Jenkins'):
        return 11
   if (x['Current_Role'] == 'Project Manager') and (x['DevOps_Tools'] == 'Other (please specify)'):
        return 12
   if (x['Current_Role'] == 'Robotics Engineer') and (x['DevOps_Tools'] in ['Kubernetes','Nginx']):
        return 13
   if (x['Current_Role'] == 'Software Engineer') and (x['DevOps_Tools'] in ['Terraform']):
        return 14
   
   else:
       return 15

# Apply the function to the DataFrame
df['devops_index'] = df.apply(lambda x: construct_devops_index(x), axis=1)

# Display the updated DataFrame
print(df)
```

```python
mlb = MultiLabelBinarizer()
```

```python
df
```

```python
df['Programming_Languages'] = df['Programming_Languages'].apply(lambda x: x if isinstance(x, list) else ([] if pd.isna(x) else x.split(",")))
```

```python
languages_encoded = mlb.fit_transform(df['Programming_Languages'])
languages_data = pd.DataFrame(languages_encoded, columns=mlb.classes_)
df = df.join(languages_data)
```

```python
df.columns
```

```python
df['Networking_Experience']=df['Networking_Experience'].map({'Yes':1,'No':0})
df['Cybersecurity_Experience']=df['Cybersecurity_Experience'].map({'Yes':1,'No':0})
df['AI_Data_Science_Experience']=df['AI_Data_Science_Experience'].map({'Yes':1,'No':0})
df['Happy_in_Current_Role']=df['Happy_in_Current_Role'].map({'Yes':1,'No':0})

df['Current_Role_Difficulty']=df['Current_Role_Difficulty'].map({'Yes':1,'No':0})
```

```python
testing=df[['year_index','Current_Role','language_index','database_index','framework_index','devops_index','Networking_Experience','Cybersecurity_Experience','AI_Data_Science_Experience','Happy_in_Current_Role','Current_Role_Difficulty']]
```

```python
testing
```

```python
# 1. Encode 'Current_Role' using Label Encoding
label_encoder = LabelEncoder()
testing['Current_Role_Encoded'] = label_encoder.fit_transform(testing['Current_Role'])

# 2. Split the data into features (X) and target (y)
y = testing[['Current_Role_Encoded']]  # Features
#X = testing[['year_index','C', 'C#', 'C++', 'Dart', 'Go','Java', 'JavaScript', 'Kotlin', 'Other (please specify)', 'PHP','Python', 'Ruby', 'Rust', 'Swift', 'TypeScript']]  # Target variable

X = testing[['year_index','language_index','database_index','framework_index','devops_index','Networking_Experience','Cybersecurity_Experience','AI_Data_Science_Experience','Happy_in_Current_Role','Current_Role_Difficulty']]  # Target variable

```

```python
X
```

```python
# 3. Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
y_train
```

```python
# 4. Train a Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# 5. Make predictions
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred.shape
```

```python
# 6. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the model: {accuracy:.2f}")
```
